```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('breast_cancer.csv')
print(data.shape)
data.head()
```

```python
# slice the data suitably and assign data-fields to Feature matrix(x) and Target matrix(Y)
data = data.iloc[:,0:32]
data.head()
```

```python
X = data.iloc[:,2:]
Y = data['diagnosis'].values
print(X[:2])
```

```python
# split the data into triining and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X ,Y ,test_size = 0.2)
```

```python
# intiantiate a model and train the model
from sklearn.tree import DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(x_train,y_train)
```

```python
# use the model for prediction
predictions = dt_classifier.predict(x_test)
prob_predictions = dt_classifier.predict_proba(x_test)
print(predictions)
print("probabiliteis of predictions :" ,prob_predictions)
```

```python
# calculate accuracy score and confusion matrix for training and testing data
from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy score for training data is :", accuracy_score(y_train,dt_classifier.predict(x_train)))
```

```python
print("Accuracy score for testing data is :", accuracy_score(y_test,dt_classifier.predict(x_test)))
```

```python
print("Confusion matrix for training data is :\n" ,confusion_matrix(y_train,dt_classifier.predict(x_train)))
```

```python
print("Confusion matrix for testing data is :\n",confusion_matrix(y_test,dt_classifier.predict(x_test)))
```
