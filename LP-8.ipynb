```python
import numpy as np
from dataviz import plot_clusters,plot_data,generate_clusters
from dbscan import DBSCAN
```

```python
# case - 1 Generate 20 random points in the range [1,100]

def generate_data(seed = None)-> np.ndarray:
    num_clusters = 4
    num_points = 20
    spread = 7
    bounds =(1,100)
    clusters = generate_clusters(num_clusters,num_points,spread,bounds,bounds,seed)
    return np.array(clusters)


clusters = generate_data(seed = 1)
plot_data(clusters)
```

```python
# applly dbscan algorithm
dbscan = DBSCAN(eps=6, min_samples=5)
dbscan.fit(clusters)
plot_clusters(clusters,dbscan.labels_,dbscan.components_)
```

```python
dbscan = DBSCAN(eps=10, min_samples=5)
dbscan.fit(clusters)
plot_clusters(clusters,dbscan.labels_,dbscan.components_)
```

```python
# case -2 generate 10 randomly generated points in the interval [1,30] [70,99]

def generate_data(bounds, seed = None)-> np.ndarray:
    num_clusters = 1
    num_points = 10
    spread = 7
    clusters = generate_clusters(num_clusters,num_points,spread,bounds,bounds,seed)
    return np.array(clusters)
```

```python
first_cluster = generate_data((1,30), seed =1)
second_cluster = generate_data((70,99),seed =1)
clusters = np.append(first_cluster,second_cluster,axis=0)
plot_data(clusters)
```

```python
dbscan = DBSCAN(eps=7, min_samples=10)
dbscan.fit(clusters)
plot_clusters(clusters,dbscan.labels_,dbscan.components_)
```

```python
# case - 3 , Generate 
# 10 random points in [1,30]
# 50 random points in [70,99]

def generate_data(num_points, bounds, seed = None) -> np.ndarray:
    num_clusters = 1
    spread = 7
    clusters = generate_clusters(num_clusters,num_points,spread,bounds,bounds,seed)
    return np.array(clusters)

first_cluster = generate_data(10,(1,30),seed = 1)
second_cluster = generate_data(50,(70,99),seed = 1)
clusters = np.append(first_cluster,second_cluster,axis=0)
plot_data(clusters)
```

```python
dbscan = DBSCAN(eps=7, min_samples=10)
dbscan.fit(clusters)
plot_clusters(clusters,dbscan.labels_,dbscan.components_)
```

```python
num_clusters = len(np.unique(dbscan.labels_))
print("Number of clusters : {}".format(num_clusters))
```
